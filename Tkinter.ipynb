```python
import tkinter as tk

window = tk.Tk()
window.title("Frame Example")
window.geometry('500x500')

#location
frame = tk.Frame(window,borderwidth=2, relief=tk.GROOVE)
frame.pack(padx=10, pady=10)

#Frame no 1 > widgets 2
textArea= tk.Text(frame, height=10, width=30)
textArea.pack()

button = tk.Button(frame,text="Enter marks")
button.pack()




frame_Mid = tk.Frame(window,borderwidth=2, relief=tk.SUNKEN)
frame_Mid.pack(padx=10, pady=10)

label2= tk.Label(frame_Mid, text="This is my frame 2", font=("Helvetica",8))
#pack to the window
label2.pack()

button2 = tk.Button(frame_Mid,text="Submit")
button2.pack()


window.mainloop()
```

```python
#radio Button
import tkinter as tk

def print_selection():
    value = var.get()
    print("Selected category is "+value)


window=tk.Tk()
window.title("Radio Button")
window.geometry('500x500')

var = tk.StringVar() #String Variable
var.set("Male") #default value


radio_button1= tk.Radiobutton(window,text="Male", variable=var, value="Male", command=print_selection)
radio_button1.pack()
radio_button2= tk.Radiobutton(window,text="Female", variable=var, value="Female", command=print_selection)
radio_button2.pack()

window.mainloop()
```

```python
#Check Box
import tkinter as tk

def print_checkbox_status():
    status= var.get()
    if status :
        print("Agreed with the conditions")
    else :
        print("You need to agree with the condition before you move a head")

window=tk.Tk()
window.title("Check box Button")
window.geometry('500x500')

var= tk.BooleanVar() # boolean variable 

checkButton= tk.Checkbutton(window, text="Agree with the condition", variable= var)
checkButton.pack()

window.mainloop()
```
