# Tkinter - File Dialog

```python
#File Handling 
# Read / Write

#File Dialog --> Selecting a file (Path)

import tkinter as tk
from tkinter import filedialog

def browsefile():
    filename= filedialog.askopenfilename() # THis will open the file DIalog and read the location of the file
    if filename :
        # Regular file open funciton
        label.config(text=filename)
        file=open(filename,"r", encoding="utf-8")
        print(file.read())
        file.close()
    else:
        print("file not found")

window = tk.Tk()
window.title("File Handilng Task")
window.geometry('500x500')

#label is used to display the content
label=tk.Label(window, text="No file selected")
label.pack()

# button is used to open the file dialog
button = tk.Button(window, text="Select a file", command = browsefile)
button.pack()

window.mainloop()
```
